<h2> Estimating Location of the Birds in King County

```python
ahumLat = []
ahumLong= []
bswiftLat = []
bswiftLong = []
brambLat = []
brambLong = []
hduckLat = []
hduckLong = []
lbuntLat = []
lbuntLong = []
sowlLat = []
sowlLong = []

def location (birdlist, latitude, longitude):
    for i in range(len(birdlist)):
        latitude.append(i)
        longitude.append(i)
        latitude[i] = birdlist[i][3]
        latitude[i] = float(latitude[i])
        longitude[i] = birdlist[i][4]
        longitude[i] = float(longitude[i])
        
location(ahum, ahumLat, ahumLong)
location(bswift, bswiftLat, bswiftLong)
location(bramb, brambLat, brambLong)
location(hduck, hduckLat, hduckLong)
location(lbunt, lbuntLat, lbuntLong)
location(sowl, sowlLat, sowlLong)

def graph (longitude, latitude, birdname):
    plt.figure(figsize=(5, 3))
    plt.grid()
    plt.scatter(longitude, latitude)
    plt.title("Location of "+birdname)
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    
graph(ahumLong, ahumLat, "Anna's Hummingbirds")
graph(bswiftLong, bswiftLat, "Black Swifts")
graph(brambLong, brambLat, "Bramblings")
graph(hduckLong, hduckLat, "Harlequin Ducks")
graph(lbuntLong, lbuntLat, "Lazuli Buntings")
graph(sowlLong, sowlLat, "Snowy Owls")
```

Interquartile Range (IQR) is a statistic found by subtracting Q3 (the third quartile) minus Q1 (the first quartile) of a data set. The IQR is significant because it tells us where the middle 50% of the numbers in the data set lie. The mathematical test for outliers is to multiply 1.5 times the IQR and add or subtract that value to the 75th and 25th percentile respectively. This calculates the upper and lower bounds. If a data value is outside of these bounds, it is mathematically an outlier of the data set. In the code I have removed all of the outliers in both latitude and longitude. 

This process has removed the statistical outliers from the data. I have not graphed the results, because as the data shows, the list lengths are not equal. This is because the latitude and longitude data for each bird sometimes have a different number of outliers. In this next, final step, I will use this data to calculate confidence intervals for the true population means of latitude and longitude for each bird.

```python
def mean_confidence_interval(birdmeasurement, data, confidence=0.95):
    a = 1.0*np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * sp.stats.t._ppf((1+confidence)/2., n-1)
    print(birdmeasurement, "confidence interval:", m-h, "to", m+h)
    print("\tMean:", m)

print("Confidence Interval Calculation Results")
mean_confidence_interval("Anna's Hummingbird Latitude", ahumLat)
mean_confidence_interval("Anna's Hummingbird Longitude", ahumLong)
mean_confidence_interval("Black Swift Latitude", bswiftLat)
mean_confidence_interval("Black Swift Longitude", bswiftLong)
mean_confidence_interval("Brambling Latitude", brambLat)
mean_confidence_interval("Brambling Longitude", brambLong)
mean_confidence_interval("Harlequin Duck Latitude", hduckLat)
mean_confidence_interval("Harlequin Duck Longitude", hduckLong)
mean_confidence_interval("Lazuli Bunting Latitude", lbuntLat)
mean_confidence_interval("Lazuli Bunting Longitude", lbuntLong)
mean_confidence_interval("Snowy Owl Latitude", sowlLat)
mean_confidence_interval("Snowy Owl Longitude", sowlLong)
```

First, I'll explain a little about what confidence intervals are. 

Say you were interested in the mean height of 18-year-old boys living in the United States. Since it is impractical to take the height all the 18-year-old boys in the United States, you just took a sample of 50 and found that the mean height was 70.2 inches, about 5'10''. This sample mean of 70.2 inches itself is of limited usefulness because it does not reveal the uncertainty associated with the estimate. You do not know how far this estimate is from the true mean height of all 18-year-old boys in the US. 

Confidence intervals provide this missing information. They are calculated using a couple complicated formulas involving mean and standard deviation and the sampling distribution, all of which I don't need to get into in this report. Confidence intervals are intervals that contain the population mean with a certain amount of confidence. For all of my calculations above, I did my calculations to be 95% confident. 

If we think about all bird observations as simply contributing to a mass sample, then a confidence interval makes sense. Bird watchers cannot possibly observe every single bird's location at every moment of the day. Therefore, when a sighting is made, it can be assumed there are more birds not currently being sighted, meaning this sighting is a sample observation of the location of all birds. 

The first confidence interval given is for the Anna's Hummingbird latitude, and it is from 47.618690615 to 47.6221372861. This means that I am 95% confident the true population mean, meaning the true average latitude of all Anna's Hummingbirds in King County, lies within this interval. For the purposes of location estimation, I have also provided the mean below the confidence interval, as the middle point between the two bounds.
